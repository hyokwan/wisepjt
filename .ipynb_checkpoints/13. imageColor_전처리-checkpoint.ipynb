```python
#!/usr/bin/env python
# coding: utf-8

# ### opencv 라이브러리 활용

# ##### 1.1라이브러리 선언

# In[2]:


import cv2
import os
import numpy as np
import matplotlib.pyplot as plt


# ##### 1.2사용함수

# In[8]:


def createFolder(directory): 
    try: 
        if not os.path.exists(directory): 
            os.makedirs(directory) 
    except OSError: 
        print ('Error: Creating directory. ' + directory)


# ##### 1.3이미지 불러오기

# In[2]:
```

```python
targetFolder = "../../문화유산_이미지세트/train406_wise382_cloudpic2_22/color/"
imgPath = targetFolder
imgList = os.listdir(imgPath) # 폴더 내 모든 파일 리스트화


# ##### 1.4 저장할 경로

# In[15]:


savePath = "../../문화유산_이미지세트/train406_wise382_cloudpic2_22/color2/"
# savePath2 = "./문화유산 이미지 자료/train_cloud/blackwhite/"

targetSaveFolder = createFolder(savePath)
# targetSaveFolder = createFolder(savePath2)

# ##### 1.5포문이용

# In[27]:


try:
    
    for i in range(0,len(imgList)):
        finalpath = imgPath + imgList[i]
        img_array = np.fromfile(finalpath, np.uint8)
        tmpImage = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
        tmpSizeImage = tmpImage
        tmpSizeImage = cv2.resize(tmpImage, (1024,1024))

        # change_Color_to_gray = cv2.cvtColor(tmpSizeImage, cv2.COLOR_BGR2GRAY)
        change_Color_to_gray=tmpSizeImage
    
        new_img_name = imgList[i]
        goingPath = savePath + imgList[i]
        extension = os.path.splitext(new_img_name)[1] # 이미지 확장자
        result, encoded_img = cv2.imencode(extension, change_Color_to_gray)
        if result:
            with open(goingPath, mode='w+b') as f:
                encoded_img.tofile(f)
                
        # (thresh, blackAndWhiteImage) = cv2.threshold(change_Color_to_gray, 127, 255, cv2.THRESH_BINARY) 
        # goingPath2 = savePath2 + imgList[i]
        # result, encoded_img = cv2.imencode(extension, blackAndWhiteImage)
        # if result:
        #     with open(goingPath2, mode='w+b') as f:
        #         encoded_img.tofile(f)
        
        print(str(i) +" 번째 성공")

except Exception as e:
    print(e)
```
